```
import matplotlib.pyplot as plt
import pandas as pd
```

```
data = pd.read_csv('push.txt', sep=" ", header=None)
data.columns = ['length', 'time']
data = data.sort_values(by='length')
plt.plot(data['length'], data['time'], 'red')

adata = pd.read_csv('pusharray.txt', sep=" ", header=None)
adata.columns = ['length', 'time']
adata = adata.sort_values(by='length')
plt.plot(adata['length'], adata['time'], 'green')
```
